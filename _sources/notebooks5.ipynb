# Clustering K-Mean Algorithm (Tugas 5)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d  

from sklearn.preprocessing import LabelEncoder
import sklearn.metrics as sm
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
url = ("https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv")
```

```python
# colnames = ['sepal.length','sepal.width','petal.length','petal.width','Class']
dt_iris = pd.read_csv(url)
```

```python
dt_iris.head(5)
```

```python
print(dt_iris.variety.value_counts())
```

Transforming the labels/species column into numbers

```python
le = LabelEncoder()
le.fit(dt_iris['variety'])
print(list(le.classes_))
dt_iris['variety'] = le.transform(dt_iris['variety'])
```

```python
print(dt_iris['variety'][0:5])
print(dt_iris['variety'][50:55])
print(dt_iris['variety'][100:105])
```

```python
dt_iris.tail()
```

Applying k-Means Algorithm

```python
iris_matrix = pd.DataFrame.to_numpy(dt_iris[['sepal.length','sepal.width','petal.length','petal.width']])
```

Define the number of clusters k = 3 and create a new model

```python
k=3
cluster_model = KMeans(n_clusters=k, random_state=10)
```

```python
print(cluster_model)
```

Compute k-Means clustering

```python
cluster_model.fit(iris_matrix)
```

```python
cluster_model.labels_
```

```python
cluster_labels = cluster_model.fit_predict(iris_matrix)
```

```python
kmeans=dt_iris # assiging iris to kmeans variableb
```

```python
kmeans.head(6)
```

Match the labels predicted by k-means with orignal labels to evaluate

```python
kmeans['pred'] = cluster_labels #adding another column to predict label and assign it k-Means model
```

```python
kmeans.head(6)
```

Computing accuracy score

```python
sm.accuracy_score(dt_iris.variety, cluster_model.labels_)
```
