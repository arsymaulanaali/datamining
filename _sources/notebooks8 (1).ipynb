```python
import numpy as np
from sklearn import preprocessing
import pandas as pd
```

```python
dataset = pd.read_csv("https://raw.githubusercontent.com/arsymaulanaali/datamining/main/credit_score1.csv")
```

```python
dataset.head()
```

```python
df = pd.DataFrame(dataset, columns = ["kode_kontrak", "pendapatan_setahun_juta", "kpr_aktif", "durasi_pinjaman_bulan", "jumlah_tanggungan", "rata_rata_overdue", "risk_rating"])
print(df)
```

```python
df2 = pd.DataFrame(dataset, columns = ["risk_rating"])
df2
```

```python
from sklearn.preprocessing import OneHotEncoder
# creating instance of one-hot-encoder
enc = OneHotEncoder(handle_unknown='ignore')
# passing bridge-types-cat column (label encoded values of bridge_types)
enc_df = pd.DataFrame(enc.fit_transform(df2[["risk_rating"]]).toarray())
# merge with main df bridge_df on key values
bridge_df = df2.join(enc_df)
bridge_df
```
