## **Import Modul**

```python
import pandas as pd
import math
```

## **Mengambil Dataset**

```python
df = pd.read_csv("https://raw.githubusercontent.com/arsymaulanaali/datamining/main/India_Menu.csv")
```

```python
df.head()
```

## **Mengambil Sampel Data**

```python
data_sample = df[["Menu Category", "Menu Items", "Energy (kCal)", "Protein (g)"]].reset_index(drop=True)
data_sample
```

## **Ukuran Kedekatan Untuk Atribut Nominal**

```python
def d1(i,j):
    P = 2
    M = 0
    for col in ["Menu Category", "Menu Items"]:
        if data_sample[col][i] == data_sample[col][j]:
            M += 1

    return (P-M)/P
```

```python
print(d1(1,2))
print(d1(1,3))
print(d1(1,4))
```

## **Jarak pada Data Numerik**

```python
def d2(i,j, h):
    result = 0
    for col in ["Energy (kCal)", "Protein (g)"]:
        result = math.pow(abs(data_sample[col][i] - data_sample[col][j]), h)
    return result**(1/h)
```

Manhattan distance

```python
print(d2(1,2,1))
print(d2(1,3,1))
print(d2(1,4,1))
```

Euclidean distance

```python
print(d2(1,2,2))
print(d2(1,3,2))
print(d2(1,4,2))
```
